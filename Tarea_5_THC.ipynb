```python
import numpy as np

# 1. Suma de matrices rectangulares
def suma_matrices(A, B):
    """Realiza la suma elemento a elemento de dos matrices del mismo tamaño.

    Args:
        A (numpy.ndarray): Primera matriz
        B (numpy.ndarray): Segunda matriz

    Returns:
        numpy.ndarray: Matriz resultante de A + B
    """
    return A + B

# 2. Multiplicación entrada a entrada (Hadamard)
def multiplicacion_elemento_a_elemento(A, B):
    """Multiplica dos matrices elemento por elemento (producto Hadamard).

    Args:
        A (numpy.ndarray): Primera matriz
        B (numpy.ndarray): Segunda matriz

    Returns:
        numpy.ndarray: Matriz resultante de A * B
    """
    return A * B

# 3. Cálculo de la traza de una matriz
def traza_matriz(A):
    """Calcula la traza (suma de elementos en la diagonal principal) de una matriz.

    Args:
        A (numpy.ndarray): Matriz de entrada (debe ser cuadrada para traza completa)

    Returns:
        float: Valor de la traza
    """
    return np.trace(A)

# 4. Transpuesta de una matriz
def transpuesta_matriz(A):
    """Calcula la matriz transpuesta (intercambia filas por columnas).

    Args:
        A (numpy.ndarray): Matriz de entrada

    Returns:
        numpy.ndarray: Matriz transpuesta
    """
    return A.T

# 5. Encontrar máximo y mínimo en una matriz
def maximo_minimo_matriz(A):
    """Encuentra el valor máximo y mínimo en una matriz.

    Args:
        A (numpy.ndarray): Matriz de entrada

    Returns:
        tuple: (maximo, minimo)
    """
    return np.max(A), np.min(A)

# 6. Suma de cuadrados de elementos
def suma_cuadrados_elementos(A):
    """Calcula la suma de los cuadrados de todos los elementos de la matriz.

    Args:
        A (numpy.ndarray): Matriz de entrada

    Returns:
        float: Suma de cuadrados
    """
    return np.sum(A**2)

# 7. Potencia de una matriz (producto matricial)
def potencia_matriz(A, n):
    """Calcula la potencia n-ésima de una matriz cuadrada (A^n).

    Args:
        A (numpy.ndarray): Matriz cuadrada
        n (int): Exponente

    Returns:
        numpy.ndarray: Matriz resultante de A^n
    """
    return np.linalg.matrix_power(A, n)

# 8. Operaciones específicas con matrices A y B
def operaciones_especificas():
    """Realiza operaciones específicas con las matrices A y B definidas en el problema."""
    A = np.array([[11, 7, -1], [-3, 5, 2], [4, 2, -2]])
    B = np.array([[1, 1, 1], [10, 3, -5], [-6, -10, -3]])

    # 2B + 3A
    resultado1 = 2*B + 3*A

    # (B + 2A)^2 (producto matricial)
    resultado2 = np.linalg.matrix_power(B + 2*A, 2)

    # A^2 + 3B
    resultado3 = np.linalg.matrix_power(A, 2) + 3*B

    print("Resultado de 2B + 3A:")
    print(resultado1)
    print("\nResultado de (B + 2A)^2:")
    print(resultado2)
    print("\nResultado de A^2 + 3B:")
    print(resultado3)


operaciones_especificas()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Familia de curvas para la ecuación diferencial
def graficar_familia_curvas():
    """Grafica la familia de curvas solución para diferentes valores de C."""
    x = np.linspace(-5, 5, 500)

    # Configuración del gráfico
    plt.figure(figsize=(10, 6))

    # Caso C > 0 (azul)
    for C in [1, 2, 3]:
        y = (x - 1) + C * np.exp(-x)
        plt.plot(x, y, 'b-', label='C > 0' if C == 1 else "")

    # Caso C = 0 (rojo)
    y = x - 1
    plt.plot(x, y, 'r-', linewidth=2, label='C = 0')

    # Caso C < 0 (verde)
    for C in [-1, -2, -3]:
        y = (x - 1) + C * np.exp(-x)
        plt.plot(x, y, 'g-', label='C < 0' if C == -1 else "")

    # Personalización del gráfico
    plt.title('Familia de curvas solución para y = (x - 1) + Ce^{-x}', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# 2. Gráfico de función racional con asíntotas
def graficar_funcion_racional():
    """Grafica la función (3x+7)/(x²-4) con sus asíntotas."""
    x = np.linspace(-10, 10, 1000)
    y = (3*x + 7) / (x**2 - 4)

    # Excluir puntos cerca de las asíntotas verticales para mejor visualización
    x = x[abs(x**2 - 4) > 0.1]  # Evita división por ~0
    y = (3*x + 7) / (x**2 - 4)

    # Configuración del gráfico
    plt.figure(figsize=(10, 6))

    # Graficar función principal
    plt.plot(x, y, 'b-', label=r'$\frac{3x + 7}{x^2 - 4}$')

    # Asíntotas verticales (x = ±2)
    plt.axvline(x=-2, color='r', linestyle='--', label='Asíntotas verticales')
    plt.axvline(x=2, color='r', linestyle='--')

    # Asíntota horizontal (y = 0)
    plt.axhline(y=0, color='g', linestyle=':', label='Asíntota horizontal')

    # Ajustar límites para mejor visualización
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)

    # Personalización
    plt.title('Gráfico de función racional con asíntotas', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

graficar_familia_curvas()
graficar_funcion_racional()
```
